```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.ticker as mtick
from decimal import *
import math
```

```python
freespace = """
-75,58	-85,58
-75,74	-85,74
-75,92	-85,92
-76,11	-86,11
-76,33	-86,33
-76,55	-86,55
-76,78	-86,78
-77,01	-87,01
-77,24	-87,24
-77,48	-87,48
-77,71	-87,71
-77,94	-87,94
-78,17	-88,17
-78,40	-88,40
-78,63	-88,63
-78,85	-88,85
""".replace(',', '.').split()
#First column is patch antenna second is isotropic

freespace = np.array([float(i) for i in freespace]).reshape(-1, 2)

fsPatch, fsIso = (freespace[:, 0], freespace[:, 1])

with open("data.raw", "r") as f:
    rawdata = f.read().replace(',','.').replace('\n', ';').split(';')

data = np.array([float(i) for i in rawdata[:-1]], dtype=np.float32).reshape(-1, 5)

d, pwDir, pwIso = (data[:-35, 0], data[:-35, 1], data[:-35, 2])

h, pwAltura = (data[:, 3], data[:, 4])
```

```python
to_dB = lambda x: 10 * np.log10(x)
to_dBV = lambda x: 2 * to_dB(np.abs(x))

Gr = lambda t: np.cos(t)**2
Gt = lambda t, iso=False: Gr(t) if iso else 10 * np.cos(t)**4
long_ond = 3e8 / 2.45e9
k = 2 * np.pi / long_ond
(hTx, hRx) = (18.5, 1.0)
L = 12
Wdisp = 0
```

```python
def modelMultipath(dh: float, iso: bool) -> float:
    """ dh: distancia horizontal
        iso: si TX isotropica
        return: potencia teorica [dBm]"""
    Rx_to_Tx = dh + (hTx - hRx)*1j
    R = np.abs(Rx_to_Tx)
    t = np.angle(Rx_to_Tx)

    Rs = np.abs(dh + (hTx + hRx)*1j)
    ts = np.arctan((hTx + hRx)/dh)

    Mp = 1 + np.sqrt((Gt(ts, iso)*Gr(ts)) / (Gt(t, iso)*Gr(t)))* -(0.5) * np.exp(-1j*k*(Rs-R))
    return Wdisp + to_dBV(long_ond/(4*np.pi*R)) + to_dB(Gt(t, iso)) + to_dB(Gr(t)) + to_dBV(Mp) - L
```

```python
horDTeo = np.arange(d[0], d[-1], (d[-1] - d[0])/1000)
pwDirTeo = (np.vectorize(lambda x: modelMultipath(x, False)))(horDTeo)
pwIsoTeo = (np.vectorize(lambda x: modelMultipath(x, True)))(horDTeo)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

def decibel_fmt(x,pos=None):
    if x==0:
        return "-inf"
    else:
        return '{:.0f}'.format(10*np.log10(np.sign(x)*x))

def decibel_fmt_minor_every_5(x,pos=None):
    if x==0:
        return "-inf"
        
    db_int = round(10*np.log10(np.sign(x)*x))
    if db_int % 5 != 0 or db_int % 10 == 0:
        return ''
    else:
        return '{:.0f}'.format(10*np.log10(np.sign(x)*x))

def decibel_fmt_minor_end_4(x,pos=None):
    if x==0:
        return "-inf"
        
    db_int = round(10*np.log10(np.sign(x)*x))
    if (db_int != -94 and db_int != -84) or db_int % 10 == 0:
        return ''
    else:
        return '{:.0f}'.format(10*np.log10(np.sign(x)*x))


linealDir = pwDir.copy()
linealIso = pwIso.copy()
linealFsPatch = fsPatch.copy()
linealFsIso = fsIso.copy()
linealDirTeo = pwDirTeo.copy()
linealIsoTeo = pwIsoTeo.copy()

# Lineal data
for lineal in (linealDir, linealIso, linealFsPatch, linealFsIso):
    for j in range(lineal.size):
        context = Context(prec=3, rounding=ROUND_UP)
        decimal_part = context.create_decimal_from_float(float(lineal[j]/10-math.floor(lineal[j]/10)))
        missing_exp = context.create_decimal_from_float(float(Decimal("10")**decimal_part))
        lineal[j] = float(Decimal("1e" + str(math.floor(lineal[j]/10)))*missing_exp)
        print(lineal[j], d[j], decimal_part, context.create_decimal_from_float(float(Decimal("10")**decimal_part)))

for lineal in (linealDirTeo, linealIsoTeo):
    for j in range(lineal.size):
        context = Context(prec=3, rounding=ROUND_UP)
        decimal_part = context.create_decimal_from_float(float(lineal[j]/10-math.floor(lineal[j]/10)))
        missing_exp = context.create_decimal_from_float(float(Decimal("10")**decimal_part))
        lineal[j] = float(Decimal("1e" + str(math.floor(lineal[j]/10)))*missing_exp)

ax1.plot(d, linealDir, label="Measures", marker="x")
ax2.plot(d, linealIso, label="Measures", marker="x")

ax1.plot(horDTeo, linealDirTeo, label="Multipath model")
ax2.plot(horDTeo, linealIsoTeo, label="Multipath model")

ax1.plot(d, linealFsPatch, label="Free space model")
ax2.plot(d, linealFsIso, label="Free space model")

ax1.set_title("Measures with array of 4 patch antennas")
ax2.set_title("Measures with isotropic antenna")

ax2.set_xlabel("Horizontal distance [m]")

for ax in (ax1, ax2):
    ax.set_ylabel("Received power [dBm]")

    ax.minorticks_on()
    ax.grid(True, which="minor", axis="y", linestyle="--")
    ax.grid(True, which="minor", axis="x", linestyle="--")
        
    ax.set_yscale("log")
    
    ax.yaxis.set_major_formatter(mtick.FuncFormatter(decibel_fmt))
    if (ax == ax1):
        ax.yaxis.set_minor_formatter(mtick.FuncFormatter(decibel_fmt_minor_every_5))
    elif (ax == ax2):
        ax.yaxis.set_minor_formatter(mtick.FuncFormatter(decibel_fmt_minor_end_4))
    
    ax.set_xlim(30, 60)
    ax.legend()
    ax.grid()

#fig.savefig("multipath_radial.png", dpi=400)
#fig.savefig("multipath_radial.pdf", bbox_inches="tight")
```
